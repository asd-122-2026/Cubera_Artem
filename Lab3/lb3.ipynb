Мета: засвоїти аналіз складності алгоритмів як технологію на прикладі
алгоритмів сортування методами включення та обміну.

зробив Цубера Артем

Теоретичні розрахунки асимптотичної складності:
1. **Сортування бульбашкою (Bubble Sort):**
   - Найгірший та середній випадки: $O(n^2)$
   - Найкращий випадок (якщо масив вже відсортований): $O(n)$
   
2. **Сортування вставлянням (Insertion Sort):**
   - Найгірший та середній випадки: $O(n^2)$
   - Найкращий випадок: $O(n)$

```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Алгоритм сортування бульбашкою
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Алгоритм сортування вставлянням
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Параметри дослідження
n_values = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]
t_bubble = []
t_insert = []

for n in n_values:
    # Генеруємо випадковий масив
    test_arr = np.random.randint(0, 10000, n)
    
    # Замір для Bubble Sort
    arr_copy = test_arr.copy()
    start = time.time()
    bubble_sort(arr_copy)
    t_bubble.append(time.time() - start)
    
    # Замір для Insertion Sort
    arr_copy = test_arr.copy()
    start = time.time()
    insertion_sort(arr_copy)
    t_insert.append(time.time() - start)
    print(f"Done for n={n}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(n_values, t_bubble, label='Bubble Sort', marker='o')
plt.plot(n_values, t_insert, label='Insertion Sort', marker='s')
plt.xlabel('Кількість елементів (n)')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння часу виконання алгоритмів сортування')
plt.legend()
plt.grid(True)
plt.show()
```

Висновок: засвоїв аналіз складності алгоритмів як технологію на прикладі алгоритмів сортування методами включення та обміну. Та переглянув та дослідив що сортування вставками ефективніше по часу чим бульбашкове сортування через те що в бульбашковом сортувані багато перестановок

Контрольні питання:
1. Що таке $O$-нотація і чим вона відрізняється від $\Omega$-нотації?Якщо просто, то $O$ — це «найгірший випадок». Вона показує максимальний час, який алгоритм може витратити на задачу. А $\Omega$ (омега) — це навпаки, «найкращий випадок» або нижня межа: швидше за цей час алгоритм точно не впорається.
2.  Яку часову складність має сортування вставлянням, якщо масив уже відсортований? Обґрунтуйте.У такому разі складність буде лінійною — $O(n)$. Це тому, що алгоритм просто один раз пройдеться по списку, побачить, що все вже стоїть правильно, і не буде нічого переставляти. Внутрішній цикл зі зсувами просто не запуститься.
3.   Що таке «ефективний алгоритм»?Це такий алгоритм, який не «їсть» зайвого. Тобто він розв'язує задачу максимально швидко і при цьому не забиває всю оперативну пам'ять комп'ютера. У програмуванні ефективними вважають ті методи, які працюють за передбачуваний час (лінійний або логарифмічний), а не ростуть до нескінченності.
4.    Що таке головний параметр завдання?Це розмір вхідних даних, від якого залежить, як довго комп'ютер буде думати. У наших лабораторних це кількість чисел у масиві ($n$). Більше чисел — більше роботи для процесора.
5. Як записати складність для $F(N) = N^3 + 7N^2 - 14N$ у $O$-нотації?Тут усе просто: ми завжди беремо тільки найбільший степінь і викидаємо все інше (коефіцієнти та менші доданки). Найбільший тут $N^3$, тому відповідь — $O(N^3)$.
