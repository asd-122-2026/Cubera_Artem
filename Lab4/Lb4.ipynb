**Завдання 1:** Реалізувати процедуру обчислення факторіалу за допомогою циклу for. Оцінити асимптотичну складність
алгоритму.

```python
def СalcFactorial(n):
    factorial=1
    for i in range(1,n+1):
        factorial *= i
    return factorial
n = int(input())
СalcFactorial(n)
```

Часова складність алгоритму становить $O(n)$, оскільки кількість операцій множення прямо залежить від значення вхідного числа $n$, яке визначає кількість ітерацій 

**Завдання 2:** Оцінити асимптотичну складність рекурсивного алгоритму обчислення факторіалу.

```python
def FactRecursion(n):
    if n == 0:
        return(1)
    return FactRecursion(n-1) * n
n = int(input())
FactRecursion(n)
```

Часова складність рекурсивного алгоритму становить $O(n)$, бо рекурсивна функція викликає сама себе таку кількість разів, яка дорівнює значенню вхідного числа $n$

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(20))
```

Часова складність становить $O(2^n)$, бо кожен виклик функції викликає дві додаткові дії, через що загальна кількість операцій подвоюється з кожним кроком збільшення значення $n$

```python
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(mylist):
    if len(mylist) < 2:
        return mylist
    middle = len(mylist) // 2
    left = mergesort(mylist[:middle])
    right = mergesort(mylist[middle:])
    return merge(left, right)
```

```python
a = [6, 5, 3, 1, 8, 7, 2, 10]
merge([2, 4, 5, 7], [1, 2, 3, 6]) 
[1, 2, 2, 3, 4, 5, 6, 7] 
mergesort(a)
```

Часова складність алгоритму  $O(n \log n)$. бо під час злиття кожен елемент обробляється рівно один раз, що дає складність $O(n)$. Але оскільки ми постійно ділимо масив навпіл, кількість таких етапів поділу становить $\log n$. Перемноживши роботу на одному етапі на кількість етапів, ми отримуємо підсумкову складність
